```python
# Mengimpor library yang diperlukan
import pandas as pd
from google_play_scraper import app, reviews, Sort
```

```python
# Menentukan ID aplikasi yang akan diambil ulasannya
app_id = "com.wb.goog.mkx"

# Mengambil semua ulasan aplikasi dari Google Play Store
scrapreview, _ = reviews(
    app_id,         # ID aplikasi yang akan di-scrape
    lang="id",      # Hanya mengambil ulasan dalam bahasa Indonesia
    country="id",   # Hanya mengambil ulasan dari pengguna di Indonesia
    sort=Sort.MOST_RELEVANT,  # Mengurutkan ulasan berdasarkan yang paling releant
    count=50000
)

# Mengonversi hasil scraping menjadi DataFrame untuk memudahkan analisis
df = pd.DataFrame(scrapreview)

# Menyimpan hasil ulasan ke dalam file CSV
df.to_csv("app_reviews.csv", index=False, encoding="utf-8")

# Menampilkan jumlah ulasan yang berhasil diambil
print(f"Jumlah ulasan yang diambil: {len(df)}")
print("Data berhasil disimpan dalam 'app_reviews.csv'")
```

```python
# Menghitung jumlah ulasan berdasarkan rating (bintang)
rating_counts = df['score'].value_counts().sort_index(ascending=False)

# Menampilkan jumlah ulasan untuk masing-masing bintang
print("\nJumlah ulasan berdasarkan rating:")
for rating in range(5, 0, -1):
    count = rating_counts.get(rating, 0)
    print(f"Bintang {rating}: {count} ulasan")
```
